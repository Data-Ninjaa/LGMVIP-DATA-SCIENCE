## Data Preprocessing

```python
df = data.filter(['Close'])
df = df.values
training_len = math.ceil( len(df) * 0.8)
print(training_len)
```

## Scaling The data

```python
scaler = MinMaxScaler(feature_range=(0,1))
scaled_df = scaler.fit_transform(df)
print(scaled_df)
```

```python
#Training Dataset
train_data = scaled_df[0:training_len,:]
print(len(train_data))
x_train=[]
y_train=[]
for i in range(60,len(train_data)):
    x_train.append(train_data[i-60:i,0])
    y_train.append(train_data[i,0])
    if i<=60:
        print(x_train)
        print(y_train)
        print()
```

```python
#Testing Dataset
test_data = scaled_df[training_len - 60: , :]
x_test=[]
y_test=df[training_len : , :]
for i in range(60, len(test_data)):
    x_test.append(test_data[i-60:i,0])
    
    
```

```python
#conversion and reshape
x_test = np.array(x_test)
x_test.shape
x_test = np.reshape(x_test,(x_test.shape[0],x_test.shape[1],1))
```

```python
#conversionand reshape
x_train, y_train = np.array(x_train),np.array(y_train)
x_train = np.reshape(x_train,(x_train.shape[0],x_train.shape[1],1))
x_train.shape
y_train.shape
```

# Building The LStm Model

```python
model = Sequential()
model.add(LSTM(units=60, return_sequences=True, input_shape=(x_train.shape[1], x_train.shape[2])))
model.add(LSTM(units=60,return_sequences=False))
model.add(Dense(1))  # Number of output features

# Compile and train the model
model.compile(optimizer='adam', loss='mean_squared_error')
model.fit(x_train, y_train, epochs=10,batch_size=32)
```

```python
#making predictions
pred = model.predict(x_test)
pred = scaler.inverse_transform(pred)
```

```python
#calculating rmse
rmse =np.sqrt(np.mean((pred - y_test) ** 2))
print("RMSE IS: ",rmse)
```

```python
#plotting actual vs predicted values
fig, ax = plt.subplots()
ax.plot(pred, '-g', label='Prediction')
ax.plot(y_test, '--r', label='Actual')
leg = ax.legend();
```
